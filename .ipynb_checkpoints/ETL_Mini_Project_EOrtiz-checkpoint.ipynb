### Create the Category and Subcategory DataFrames
---
**Create a Category DataFrame that has the following columns:**
- A "category_id" column that is numbered sequential form 1 to the length of the number of unique categories.

Export the DataFrame as a `category.csv` CSV file.

```python
# Get the crowdfunding_info_df columns.
crowdfunding_info_df.columns
```

```python
# Assign the category and subcategory values to 'category' and 'subcategory' columns.
crowdfunding_info_df[['category', 'subcategory']] = crowdfunding_info_df['category & sub-category'].str.split('/', expand=True)
```

```python
#Print crowdfunding_info_df after the split
crowdfunding_info_df.head()
```

```python
# Basic Descriptive statistics
crowdfunding_info_df[["goal", "pledged", "backers_count"]].describe()
```

```python
#Count outcome
outcome_counts=crowdfunding_info_df['outcome'].value_counts()
outcome_counts
```

```python
# Outcome distribution
outcome_counts.plot(kind='bar')

plt.title('Outcome Distribution')
plt.xlabel('Outcome')
plt.ylabel('Number of Campaigns')
plt.show()
```

```python
# Scatter plot of goal vs pledged
plt.figure(figsize=(10, 6))
plt.scatter(crowdfunding_info_df['goal'], crowdfunding_info_df['pledged'], alpha=0.5)
plt.title('Goal vs Pledged Amounts')
plt.xlabel('Goal Amount')
plt.ylabel('Pledged Amount')
plt.show()
```

```python
# Scatter plot of backers count vs pledged
plt.figure(figsize=(10, 6))
plt.scatter(crowdfunding_info_df['backers_count'], crowdfunding_info_df['pledged'], alpha=0.5)
plt.title('Backers Count vs Pledged Amount')
plt.xlabel('Backers Count')
plt.ylabel('Pledged Amount')
plt.show()
```

```python
#Count country
country_counts = crowdfunding_info_df['country'].value_counts()
country_counts
```

```python
# Country breakdown bar plot
country_counts.plot(kind='bar')
plt.title('Campaigns by Country')
plt.xlabel('Country')
plt.ylabel('Number of Campaigns')
plt.xticks(rotation=45)
plt.show()
```

```python
#Success rate by category count
category_success_rate = crowdfunding_info_df.groupby('category')['outcome'].value_counts(normalize=True).unstack()
category_success_rate 
```

```python
# Success rate by category
category_success_rate.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Success Rate by Category')
plt.xlabel('Category')
plt.ylabel('Proportion of Campaigns')
plt.xticks(rotation=45)
plt.show()
```

```python
#Get the unique categories and subcategories.
categories = crowdfunding_info_df['category'].unique()
subcategories = crowdfunding_info_df['subcategory'].unique()
```

```python
#Print the unique categories and subcategories
print(categories)
print(subcategories)
```

```python
# Get the number of distinct values in the categories and subcategories lists.
print(len(categories))
print(len(subcategories))
```

```python
# Creating numpy arrays from 1-9 for the categories and 1-24 for the subcategories.
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)

print(category_ids)
print(subcategory_ids)
```

```python
# Use a list comprehension to add "cat" to each category_id
cat_ids = [f'cat{cat_id}' for cat_id in category_ids]

# Use a list comprehension to add "subcat" to each subcategory_id.  
scat_ids = [f'subcat{subcat_id}' for subcat_id in subcategory_ids]
print(cat_ids)
print(scat_ids)
```

```python
# Creating a category DataFrame with the category_id array as the category_id and categories list as the category name.
category_df = pd.DataFrame({
    'category_id': cat_ids,
    'category': categories
})
# Creating a category DataFrame with the subcategory_id array as the subcategory_id and subcategories list as the subcategory name. 
subcategory_df = pd.DataFrame({
    'subcategory_id': scat_ids,
    'subcategory': subcategories
})
```

```python
#Printing category_df
category_df
```

```python
#Printing subcategory_df
subcategory_df
```

```python
# Exporting categories_df and subcategories_df as CSV files.
category_df.to_csv("Resources/category.csv", index=False)

subcategory_df.to_csv("Resources/subcategory.csv", index=False)
```

### Campaign DataFrame
----
**Create a Campaign DataFrame that has the following columns:**
- The "cf_id" column.
- The "contact_id" column.
- The company_name column.
- The "blurb" column is renamed as "description."
- The "goal" column.
- The "goal" column is converted to a `float` datatype.
- The "pledged" column is converted to a `float` datatype. 
- The "backers_count" column. 
- The "country" column.
- The "currency" column.
- The "launched_at" column is renamed as "launch_date" and converted to a datetime format. 
- The "deadline" column is renamed as "end_date" and converted to a datetime format.
- The "category_id" with the unique number matching the category_id from the category DataFrame. 
- The "subcategory_id" with the unique number matching the subcategory_id from the subcategory DataFrame.
- And, create a column that contains the unique four-digit contact ID number from the `contact.xlsx` file.
 

Then export the DataFrame as a `campaign.csv` CSV file.


```python
# Creating a copy of the crowdfunding_info_df DataFrame name campaign_df. 
campaign_df = crowdfunding_info_df.copy()
campaign_df.head()
```

```python
# Renaming the blurb, launched_at, and deadline columns.
campaign_df.rename(columns={
    'blurb': 'description',
    'launched_at': 'launch_date',
    'deadline': 'end_date'
}, inplace=True)
campaign_df.head()
```

```python
#Converting the 'launch_date' and 'end_date' columns to datetime format
campaign_df['launch_date'] = campaign_df['launch_date'].apply(lambda x: dt.utcfromtimestamp(x).date())
campaign_df['end_date'] = campaign_df['end_date'].apply(lambda x: dt.utcfromtimestamp(x).date())
```

```python
#Printing campaign_df
campaign_df.head()
```

```python
# Converting the goal and pledged columns to a `float` data type.
campaign_df['goal'] = campaign_df['goal'].astype(float)
campaign_df['pledged'] = campaign_df['pledged'].astype(float)
```

```python
#Priting campaign_df
campaign_df.head()
```

```python
#Printing the data info
campaign_df.info()
```

```python
## Checking the datatypes
campaign_df.dtypes
```

```python
#Printing the DF
campaign_df.head()
```

```python
# Merging the campaign_df with the category_df on the "category" column and 
# the subcategory_df on the "subcategory" column.
campaign_merged_df = campaign_df.merge(category_df, how='left', left_on='category', right_on='category')
campaign_merged_df = campaign_merged_df.merge(subcategory_df, how='left', left_on='subcategory', right_on='subcategory')
campaign_merged_df.tail(10)
```

```python
#Dropping unwanted columns (category , subcategory columns and the original 'category' and 'subcategory')
campaign_cleaned = campaign_merged_df.drop(columns=['staff_pick','spotlight','category', 'subcategory', 'category & sub-category'])
campaign_cleaned.head()
```

```python
# Exporting the DataFrame as a CSV file. 
campaign_cleaned.to_csv("Resources/campaign.csv", index=False)
```

### Create the Contacts DataFrame 
---
**Create a Contacts DataFrame that has the following columns:**
- A column named "contact_id"  that contains the unique number of the contact person.
- A column named "first_name" that contains the first name of the contact person.
- A column named "last_name" that contains the first name of the contact person.
- A column named "email" that contains the email address of the contact person

Then export the DataFrame as a `contacts.csv` CSV file.

### Option 2: Use regex to create the contacts DataFrame.

```python
#printing contact_info_df
contact_info_df.head()
```

```python
# Extracting the four-digit contact ID number.
contact_info_df['contact_id'] = contact_info_df['contact_info'].str.extract(r'"contact_id": (\d{4})')
contact_info_df.head()
```

```python
# Checking the datatypes.
contact_info_df.info()
```

```python
# Converting the "contact_id" column to an int64 data type.
contact_info_df['contact_id'] = contact_info_df['contact_id'].astype('int64')
```

```python
# Checking the datatypes.
contact_info_df.info()
```

```python
# Extracting the name of the contact and add it to a new column.
contact_info_df['name'] = contact_info_df['contact_info'].str.extract(r'"name": "([^"]+)"')
contact_info_df.head(10)
```

```python
# Extracting the email from the contacts and add the values to a new column.
contact_info_df['email'] = contact_info_df['contact_info'].str.extract(r'"email": "([^"]+)"')
contact_info_df.head(10)
```

```python
# Creating a copy of the contact_info_df with the 'contact_id', 'name', 'email' columns.
contact_info_copy = contact_info_df[['contact_id', 'name', 'email']].copy()
contact_info_copy.head(10)
```

```python
# Creating a "first"name" and "last_name" column with the first and last names from the "name" column. 
contact_info_copy[['first_name', 'last_name']] = contact_info_copy['name'].str.split(' ', n=1, expand=True)
# Dropping the contact_name column
contact_info_copy = contact_info_copy.drop('name', axis=1)
contact_info_copy.head()
```

```python
# Reordering the columns
contact_info_copy = contact_info_copy[['contact_id', 'first_name', 'last_name', 'email']]
contact_info_copy.head()
```

```python
# Checking the datatypes.
contact_info_copy.info()
```

```python
# Exporting the DataFrame as a CSV file. 
contact_info_copy.to_csv("Resources/contacts.csv", encoding='utf8', index=False)
```
